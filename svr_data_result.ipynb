```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

from sklearn.svm import LinearSVR
```

```python
data = pd.read_csv('gm_data.csv',index_col='Unnamed: 0')
data_d = data
```

```python
def norm(d):
    r = (d - np.mean(d))/np.std(d)
    return r
```

```python
data = norm(data)
data
```

```python
clf = LinearSVR()
```

```python
# data.index
data_train = data.loc[range(1994,2014)]
```

```python
data_train
```

```python
x_train = data_train.iloc[:,:-1]
y_train = data_train['y']
```

```python
clf.fit(x_train,y_train)
```

```python
x = (data.iloc[:,:-1] - data.iloc[:,:-1].mean()) / data.iloc[:,:-1].std()
```

```python
data_d['pred'] = clf.predict(x) * data_d['y'].std() + data_d['y'].mean()
data_d
```

```python
fig = plt.subplots(figsize=(10,8))
plt.plot(data_d[['y','pred']])
```

```python
def mape(y_true,y_pred):  # 平均绝对百分误差  越接近于0 说明效果越好
    sum_err = abs((y_true-y_pred)/y_true).sum()
    r = sum_err/len(y_true)
    return r * 1.0
```

```python
mape = mape(data_d['y'],data_d['pred'])
mape
```
